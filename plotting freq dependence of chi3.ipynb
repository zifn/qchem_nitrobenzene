```python
#Note Boyd References a value of 5.7 * 10^-20 likely measured at 1064 nm or 532 nm

from os import path
from copy import deepcopy

from IPython.display import Math, Latex
import sympy as sp
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sympy import init_printing

import get_effective_freq_dependence_from_raw_output as main

init_printing()
```

```python
def eV_and_hartree(df):
    df_ev = deepcopy(df)
    df_hartree = deepcopy(df)
    
    df_ev.index = df_ev['freq (eV)']
    df_hartree = df_hartree.drop('freq (eV)', 1)
    
    df_hartree.index = df_hartree['freq (Hartree)']
    df_ev = df_ev.drop('freq (Hartree)', 1)
    return df_ev, df_hartree

def plot_data_frames(output_df, good_df, descripter):
    display(output_df)
    display(good_df)
 
    plt.plot(output_df['freq (Hartree)'], output_df['gamma effective'], 'bo')
    plt.plot(good_df['freq (Hartree)'], good_df['gamma effective'], 'go')
    y_label = "$\gamma_{\t_{eff}}$ (au.)"
    x_label = "freq (Hartree)"
    plt.title("Calculated Rotationally Averaged {0}\n{1}\nDunning Double Zeta".format("$\gamma_{\t_{effective}}$", descripter), fontsize=24, y=1.08)
    plt.tick_params(labelsize=13)
    plt.ylabel(y_label, fontsize=20)
    plt.xlabel(x_label, fontsize=20)
    plt.show()

    plt.plot(output_df['freq (eV)'], output_df['gamma effective'], 'bo')
    plt.plot(good_df['freq (eV)'], good_df['gamma effective'], 'go')
    y_label = "$\gamma_{\t_{eff}}$ (au)"
    x_label = "freq (eV)"
    plt.title("Calculated Rotationally Averaged {0}\n{1}\nDunning Double Zeta".format("$\gamma_{\t_{effective}}$", descripter), fontsize=24, y=1.08)
    plt.tick_params(labelsize=13)
    plt.ylabel(y_label, fontsize=20)
    plt.xlabel(x_label, fontsize=20)
    plt.show()

    plt.plot(output_df['freq (Hartree)'], output_df['chi3 effective'], 'bo')
    plt.plot(good_df['freq (Hartree)'], good_df['chi3 effective'], 'go')
    plt.plot([0.0428], [5.7*10**(-20)], 'ro')
    y_label = "$\chi^{(3)}_{\t_{eff}}}$ (SI)"
    x_label = "freq (Hartree)"
    plt.title("Calculated Rotationally Averaged {0}\nWith Lorentz Lorenz Local Field Correction\n{1}\nDunning Double Zeta".format("$\chi^{(3)}_{\t_{eff}}}$", descripter), fontsize=24, y=1.08)
    plt.tick_params(labelsize=13)
    plt.ylabel(y_label, fontsize=20)
    plt.xlabel(x_label, fontsize=20)
    plt.show()

    plt.plot(output_df['freq (eV)'], output_df['chi3 effective'], 'b+', label='raw output')
    plt.plot(good_df['freq (eV)'], good_df['chi3 effective'], 'gx', label='no warning')
    plt.plot([1.1653], [5.7*10**(-20)], 'ro', label="reported ground state")
    x_label = "freq (eV)"
    y_label = "$\chi^{(3)}_{\t_{eff}}}$ (SI)"
    plt.title("Calculated Rotationally Averaged {0}\nWith Lorentz Lorenz Local Field Correction\n{1}\nDunning Double Zeta".format("$\chi^{(3)}_{\t_{eff}}}$", descripter), fontsize=24, y=1.08)
    plt.tick_params(labelsize=13)
    plt.ylabel(y_label, fontsize=20)
    plt.xlabel(x_label, fontsize=20)
    plt.legend()
    plt.show()
```

```python
output_df_state2 = main.main('main_calc_output_files', 'effective_chi3_freq_response_NBopt_dunningZ_2.csv', True)
good_df_state2 = output_df_state2[output_df_state2.warning_flag != True]

output_df_state1 = main.main('NB_ground_state_freq_dependence', 'effective_chi3_freq_response_NBopt_dunningZ_2.csv', True)
good_df_state1 = output_df_state1[output_df_state1.warning_flag != True]

```

```python
plot_data_frames(output_df_state2, good_df_state2, "Excited State")
```

```python
plot_data_frames(output_df_state1, good_df_state1, "Ground State")
```

```python
output_df_state2_ev, output_df_state2_hrt = eV_and_hartree(output_df_state2)
output_df_state1_ev, output_df_state1_hrt = eV_and_hartree(output_df_state1)

#display(output_df_state2_ev)
#display(output_df_state2_hrt)

output_df_delta_ev = output_df_state2_ev - output_df_state1_ev
output_df_delta_hrt = output_df_state2_hrt - output_df_state1_hrt

#display(output_df_delta_ev)
#display(output_df_delta_hrt)

output_df_delta = deepcopy(output_df_delta_ev)
output_df_delta['freq (eV)'] = output_df_delta_ev.index
output_df_delta['freq (Hartree)'] = output_df_delta_hrt.index
#display(output_df_delta)

output_df_delta.warning_flag = output_df_state2_ev.warning_flag|output_df_state1_ev.warning_flag
display(output_df_delta)
output_df_delta = output_df_delta.dropna(0)
display(output_df_delta)
good_df_delta = output_df_delta[output_df_delta.warning_flag != True]


```

```python
plot_data_frames(output_df_delta, good_df_delta, "Difference between states")
```
